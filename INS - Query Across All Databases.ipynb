```python
# ENTER SQL CREDENTIALS HERE
server = 'ins-prod-sql-usw.database.windows.net'
username = 'rngo'
password = ''
database = ''

df = pd.DataFrame()
for k in range(19):
    if len(str(k+1)) > 1: 
        database = 'insellerateCrm-A0'+str(k+1)
    else: 
        database = 'insellerateCrm-A00'+str(k+1)
        
    cnxn = pyodbc.connect('DRIVER={SQL Server};SERVER='+server+';DATABASE='+database+';UID='+username+';PWD='+ password)
    cursor = cnxn.cursor()
    
    # ENTER QUERY TO RUN ACROSS ALL DB HERE
    query = '''
    select o.name, o.orgid, count(ws.webhooksettingid) from webhooksetting ws (nolock)
left join webhooksettinghistory wsh (nolock) on wsh.webhooksettingid = ws.webhooksettingid and wsh.versionid = ws.versionid
left join org o (nolock) on o.orgid = ws.orgid 
where wsh.createddate >= '2021-11-01' and ws.isactive = 1
group by o.name, o.orgid
    '''
    res = pd.read_sql(query, cnxn)
    df = df.append(res)


display(df)
df.to_clipboard(excel=True) # AUTO-COPIES-- ENTER CTRL-V TO PASTE RESULTS ONTO EXCEL SPREADSHEET
```
