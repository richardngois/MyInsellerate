```python
acctype = widgets.Dropdown(
        options=[
            'All', 
            'Campaign Lead Post',
            'YTel Callback Api',
            'General Integration Api',
            'SendGrid Webhook Api',
            'Call Center Api',
            'Encompass Api'],
        value='Campaign Lead Post',
        description='Access Type:',
        disabled=False,
    )
display(acctype)
```

```python
# specify paths to access type dropdown options
if acctype.value == 'All':
    path = '//*[@id="wrapper"]/ins-app-container/div[2]/div[1]/div/ui-view/div/div[2]/div/div/div/div[2]/form/div/div[1]/div[2]/div/ins-dropdown/div/div/div/div/div[1]/li[1]'
elif acctype.value == 'Campaign Lead Post':
    path = '//*[@id="wrapper"]/ins-app-container/div[2]/div[1]/div/ui-view/div/div[2]/div/div/div/div[2]/form/div/div[1]/div[2]/div/ins-dropdown/div/div/div/div/div[1]/li[2]'
elif acctype.value == 'YTel Callback Api':
    path = '//*[@id="wrapper"]/ins-app-container/div[2]/div[1]/div/ui-view/div/div[2]/div/div/div/div[2]/form/div/div[1]/div[2]/div/ins-dropdown/div/div/div/div/div[1]/li[3]'
elif acctype.value == 'General Integration Api':
    path = '//*[@id="wrapper"]/ins-app-container/div[2]/div[1]/div/ui-view/div/div[2]/div/div/div/div[2]/form/div/div[1]/div[2]/div/ins-dropdown/div/div/div/div/div[1]/li[4]'
elif acctype.value == 'SendGrid Webhook Api':
    path = '//*[@id="wrapper"]/ins-app-container/div[2]/div[1]/div/ui-view/div/div[2]/div/div/div/div[2]/form/div/div[1]/div[2]/div/ins-dropdown/div/div/div/div/div[1]/li[5]'
elif acctype.value == 'Call Center Api':
    path = '//*[@id="wrapper"]/ins-app-container/div[2]/div[1]/div/ui-view/div/div[2]/div/div/div/div[2]/form/div/div[1]/div[2]/div/ins-dropdown/div/div/div/div/div[1]/li[6]'
else:
    path = '//*[@id="wrapper"]/ins-app-container/div[2]/div[1]/div/ui-view/div/div[2]/div/div/div/div[2]/form/div/div[1]/div[2]/div/ins-dropdown/div/div/div/div/div[1]/li[7]'
    
# specify paths to dropdown fields and button cmds on INS page
mapping_tab = driver.find_element(By.XPATH,'//*[@id="wrapper"]/ins-app-container/div[2]/div[1]/div/ui-view/div/div[2]/div/div/ul/li[2]/a')
ips_dropdown = driver.find_element(By.XPATH,'//*[@id="ipMap"]')
type_dropdown = driver.find_element(By.XPATH,'//*[@id="type"]')
type_add = driver.find_element(By.XPATH,'//*[@id="wrapper"]/ins-app-container/div[2]/div[1]/div/ui-view/div/div[2]/div/div/div/div[2]/form/div/div[1]/div[3]/div/button')

# loop through Ips to be mapped to the selected access type
for k in range(len(ips)):
    if k == 0:
        print("Starting access-type mapping...")
        
    try:
        # scroll to top of page to click on the mapping tab
        driver.execute_script("return arguments[0].scrollIntoView(true);",top)
        mapping_tab.click()
    except:
        # scroll to top of page (at least where the mapping tab is visible) to click on the mapping tab
        driver.execute_script("return arguments[0].scrollIntoView(true);",mapping_tab)
        mapping_tab.click()
    
    # select the Ip Address dropdown field
    ips_dropdown.click()

    # each iter will select an Ip from your list 
    ips_select = driver.find_element(By.XPATH,'//*[@id="wrapper"]/ins-app-container/div[2]/div[1]/div/ui-view/div/div[2]/div/div/div/div[2]/form/div/div[1]/div[1]/div/ins-dropdown/div/div/div/div/div[1]/li/*[contains( text(), "%s")]' % ips[k])
    driver.execute_script("return arguments[0].scrollIntoView(true);",ips_select) # scroll through dropdown options where value is visible
    ips_select.click()

    # select the Access Type dropdown field
    driver.execute_script("return arguments[0].scrollIntoView(true);",top) # scroll to top where cmd is visible
    type_dropdown.click()

    # select the access type selected from the previous cell
    type_select = driver.find_element(By.XPATH,path)
    driver.execute_script("return arguments[0].scrollIntoView(true);",top) # scroll to top where cmd is visible
    type_select.click()

    # map the Ip from the current iter to the selected access type
    driver.execute_script("return arguments[0].scrollIntoView(true);",top) # scroll to top where cmd is visible
    type_add.click()

    # close duplicate Ip popup window if alerted 
    # allow 2.4 sec delay before and after for the popup window to receive the response
    time.sleep(2.4)
    try:
        driver.switch_to.alert.accept()
        print(ips[k] + " already has access type: " + acctype.value + " ...")
    except:
        print("Access type has been added for " + ips[k] + " ...")     
    time.sleep(2.4)

    # notify notebook user when task is complete
    if k == len(ips)-1:
        print("Complete")
```
